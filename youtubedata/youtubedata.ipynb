```python
df.reset_index(drop=True, inplace=True)
df.tail()
```

```python
plt.figure(figsize=(9,4))
sns.histplot(df['View_count'],bins=20,kde=True)
plt.title('Distribution of View Count')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Like_count'], bins=30, kde=True, color='green')
plt.title('Distribution of Like Count')
plt.show()
```

```python
# Comment Count
plt.figure(figsize=(8, 4))
sns.histplot(df['Comment_count'], bins=30, kde=True, color='orange')
plt.title('Distribution of Comment Count')
plt.show()
```

```python
plt.figure(figsize=(8, 2))
sns.boxplot(x=df['View_count'])
plt.title('Boxplot of View Count')
plt.show()
```

```python
# Like Count
plt.figure(figsize=(8, 2))
sns.boxplot(x=df['Like_count'], color='green')
plt.title('Boxplot of Like Count')
plt.show()
```

```python
# Comment Count
plt.figure(figsize=(8, 2))
sns.boxplot(x=df['Comment_count'], color='orange')
plt.title('Boxplot of Comment Count')
plt.show()
```

```python
df.skew(numeric_only=True)
```

```python
df.kurtosis(numeric_only=True)
```

```python
sns.heatmap(df.isnull(), cbar=True, cmap='viridis')
```

```python
Q1 = df['View_count'].quantile(0.25)
Q3 = df['View_count'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['View_count'] < (Q1 - 1.5 * IQR)) | (df['View_count'] > (Q3 + 1.5 * IQR))]
print(outliers)
```

```python
outliers[['Title', 'View_count','Like_count','Comment_count']].head(10)
```

```python
sns.pairplot(df[['View_count', 'Like_count', 'Comment_count']])
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
```
